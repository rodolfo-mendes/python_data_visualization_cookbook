```python
# sample data
x = [1,2,3,4]
y = [5,4,3,2]
```

```python
plt.figure('Sample Charts')
plt.subplot(2, 3, 1) # divide subplots into a 2x3 grid and select cell #1
plt.plot(x,y)
```

```python
plt.subplot(2, 3, 2) # select cell #2
plt.bar(x, y) #plot a vertical bar chart
```

```python
plt.subplot(2, 3, 3) # select cell #3
plt.barh(x, y) #plot a horizontal bar chart
```

```python
# create stacked bar chart
plt.subplot(2, 3, 4)
plt.bar(x, y)

y1 = [7, 8, 5, 3]
plt.bar(x, y1, bottom=y)
```

```python
# boxplot
plt.subplot(2, 3, 5)
plt.boxplot(x)
```

```python
# scatter plot
plt.subplot(2, 3, 6, )
plt.scatter(x, y)
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)

y = np.cos(x)
y1 = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.plot(x, y1)

plt.title(r'Functions $\sin$ and $\cos$')
plt.xlim(-3.0, 3.0)
plt.ylim(-1.0, 1.0)

(plt.xticks(
    [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
    [r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', r'$\frac{\pi}{2}$', r'$\pi$']))

plt.yticks([-1, 0, 1], [r'$-1$',r'$0$',r'$+1$'])
plt.show()
```
