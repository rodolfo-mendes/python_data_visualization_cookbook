```python
# sample data
x = [1,2,3,4]
y = [5,4,3,2]
```

```python
plt.figure('Sample Charts')
plt.subplot(2, 3, 1) # divide subplots into a 2x3 grid and select cell #1
plt.plot(x,y)
```

```python
plt.subplot(2, 3, 2) # select cell #2
plt.bar(x, y) #plot a vertical bar chart
```

```python
plt.subplot(2, 3, 3) # select cell #3
plt.barh(x, y) #plot a horizontal bar chart
```

```python
# create stacked bar chart
plt.subplot(2, 3, 4)
plt.bar(x, y)

y1 = [7, 8, 5, 3]
plt.bar(x, y1, bottom=y)
```

```python
# boxplot
plt.subplot(2, 3, 5)
plt.boxplot(x)
```

```python
# scatter plot
plt.subplot(2, 3, 6, )
plt.scatter(x, y)
```
