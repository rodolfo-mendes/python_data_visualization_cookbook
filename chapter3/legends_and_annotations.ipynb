```python
from matplotlib import pyplot as plt

import numpy as np

%matplotlib notebook
```

```python
#generate different normal distributions
x1 = np.random.normal(30, 3, 100)
x2 = np.random.normal(20, 2, 100)
x3 = np.random.normal(10, 3, 100)
```

```python
# plot them
_ = plt.plot(x1, label = 'plot')
_ = plt.plot(x2, label = '2nd plot')
_ = plt.plot(x3, label = '3rd plot')
```

```python
# generate a legend box
_ = plt.legend(bbox_to_anchor=(0, 1.02, 1, .102), loc=3, ncol=3, mode='expand', borderaxespad=0)
```

```python
# annotate an important value 
_ = plt.annotate('Important value', (55, 20), xycoords='data', xytext=(5, 38), arrowprops={'arrowstyle':'->'})
```

```python
# testing different locations
plt.figure()
plt.plot(x1, label = 'plot')
plt.plot(x2, label = '2nd plot')
plt.plot(x3, label = '3rd plot')
plt.legend(loc='upper left')
```

```python
plt.figure()
plt.plot(x1, label = 'plot')
plt.plot(x2, label = '2nd plot')
plt.plot(x3, label = '3rd plot')
plt.legend(loc='lower right')
```

```python
plt.figure()
plt.plot(x1, label = 'plot')
plt.plot(x2, label = '2nd plot')
plt.plot(x3, label = '3rd plot')

plt.legend(loc=3) #puts the legend at the lower left position
```

```python
plt.legend(loc=3, ncol=3) # applies 3 column layout to legends
```

```python
plt.legend(loc=3, ncol=3, mode='expand') # expands the legend to the length of axis
```

```python
plt.legend(loc=3, ncol=3, mode='expand', borderaxespad=0) # suppres the pad between the box and the axis
```

```python
# bbox_to_anchor places the legend at an arbitray position

x_pos = 0
y_pos = 1.02
width = 1
heigth = .102

plt.legend(bbox_to_anchor=(x_pos, y_pos, width, heigth), loc=3, ncol=3, mode='expand', borderaxespad=0)
```
