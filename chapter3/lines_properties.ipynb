```python
# testing for different line styles, colors and markers

x = np.arange(0, 10)

plt.figure()
l1 = plt.plot(x, 0.1 * x, color='yellowgreen', alpha=0.6, linestyle='-', marker='o')
l2 = plt.plot(x, 0.2 * x, color='maroon', alpha=0.6, linestyle='--', marker='D')
l3 = plt.plot(x, 0.3 * x, color='royalblue', alpha=0.6, linestyle='-.', marker='*', linewidth=3)
l4 = plt.plot(x, 0.5 * x, color='darkorange', alpha=0.6, linestyle=':', marker='v')
```

```python
# testing different cap and join styles
x = np.arange(0, 10)
y = np.array( [0, 10, 20, 30, 25, 15, 8, 17, 31, 40] )

plt.figure()
l1, = plt.plot(x, y, linewidth = 7) #defaults: projecting and round
l2 = plt.plot(x, y + 10 , linewidth = 7, solid_capstyle='projecting', solid_joinstyle='round')
l3 = plt.plot(x, y + 20, linewidth = 7, solid_capstyle='round', solid_joinstyle='bevel')
l4 = plt.plot(x, y + 30, linewidth = 7, solid_capstyle='butt', solid_joinstyle='miter')
```
