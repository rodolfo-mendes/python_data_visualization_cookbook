```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
mu = 100
sigma = 15
x = np.random.normal(mu, sigma, 1000)
```

```python
ax = plt.gca()
```

```python
_ = ax.hist(x, bins=35, color='tab:red', alpha=0.3)
```

```python
ax.set_xlabel('Values')
ax.set_ylabel('Frequency')
```

```python
_ = ax.set_title(r'$\mathrm{Histogram:}\ \mu=%d,\ \sigma=%d $' % (mu, sigma))
```

### The same data, displayed in horizontal

```python
plt.figure()
_ = plt.hist(x, bins=35, color='tab:red', alpha=0.3, orientation='horizontal')
```

### Other types of histogram

```python
plt.figure()
_ = plt.hist(x, bins=35, color='tab:red', alpha=0.3, histtype='step')
```

```python
plt.figure()
_ = plt.hist(x, bins=35, color='tab:red', alpha=0.3, histtype='stepfilled')
```

### Stacked bar

```python
# more data
x1 = np.random.normal(mu, sigma, 1000)
```

```python
plt.figure()
plt.hist([x, x1], bins=35, stacked=True)
#plt.hist(x1, bins=35, stacked=True)
```
